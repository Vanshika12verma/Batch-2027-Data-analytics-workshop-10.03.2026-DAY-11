```python
from sklearn.datasets import load_iris
import numpy as np

# Load iris dataset
iris = load_iris()

data = iris.data
target = iris.target
species = iris.target_names

# Convert to list of lists
data_list = data.tolist()

print("Mean and Standard Deviation for each column:\n")

columns = ["Sepal Length", "Sepal Width", "Petal Length", "Petal Width"]

for i in range(4):
    col = [row[i] for row in data_list]
    print(columns[i])
    print("Mean:", np.mean(col))
    print("Standard Deviation:", np.std(col))
    print()

print("\nMean and Standard Deviation for each species:\n")

for s in range(len(species)):
    species_data = [data_list[i] for i in range(len(data_list)) if target[i] == s]

    print("Species:", species[s])

    for j in range(4):
        col = [row[j] for row in species_data]
        print(columns[j])
        print("Mean:", np.mean(col))
        print("Standard Deviation:", np.std(col))

    print()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from mpl_toolkits.mplot3d import Axes3D

# Load dataset
iris = load_iris()

data = iris.data.tolist()      # list of lists
target = iris.target
species = iris.target_names

columns = ["Sepal Length", "Sepal Width", "Petal Length", "Petal Width"]

# -------------------------------
# Mean and Standard Deviation
# -------------------------------

print("Mean and Standard Deviation (Overall)\n")

for i in range(4):
    col = [row[i] for row in data]
    print(columns[i])
    print("Mean:", np.mean(col))
    print("Std Dev:", np.std(col))
    print()

# -------------------------------
# Mean and SD for each species
# -------------------------------

print("\nMean and Standard Deviation for each Species\n")

for s in range(3):

    species_data = [data[i] for i in range(len(data)) if target[i] == s]

    print("Species:", species[s])

    for j in range(4):
        col = [row[j] for row in species_data]

        print(columns[j])
        print("Mean:", np.mean(col))
        print("Std Dev:", np.std(col))

    print()

# -------------------------------
# Box Plot
# -------------------------------

plt.figure(figsize=(8,6))
plt.boxplot(np.array(data))
plt.xticks([1,2,3,4], columns)
plt.title("Box Plot of Iris Features")
plt.show()

# -------------------------------
# Scatter Plot
# -------------------------------

plt.figure(figsize=(8,6))
plt.scatter([row[0] for row in data], [row[2] for row in data], c=target)
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Scatter Plot of Iris Data")
plt.show()

# -------------------------------
# Outliers using Box Plot (IQR)
# -------------------------------

print("\nOutliers using IQR method\n")

for i in range(4):

    col = np.array([row[i] for row in data])

    Q1 = np.percentile(col, 25)
    Q3 = np.percentile(col, 75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR

    outliers = col[(col < lower) | (col > upper)]

    print(columns[i], "Outliers:", outliers)

# -------------------------------
# Histogram
# -------------------------------

plt.figure(figsize=(8,6))
plt.hist([row[0] for row in data], bins=10, color="skyblue")
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()

# -------------------------------
# Bar Chart
# -------------------------------

counts = [list(target).count(i) for i in range(3)]

plt.figure(figsize=(6,6))
plt.bar(species, counts, color=["red","green","blue"])
plt.title("Bar Chart of Iris Species Count")
plt.show()

# -------------------------------
# Pie Chart
# -------------------------------

plt.figure(figsize=(6,6))
plt.pie(counts, labels=species, autopct='%1.1f%%', colors=["gold","lightgreen","lightblue"])
plt.title("Pie Chart of Iris Species")
plt.show()

# -------------------------------
# 3D Histogram
# -------------------------------

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.array([row[0] for row in data])
y = np.array([row[1] for row in data])

hist, xedges, yedges = np.histogram2d(x, y, bins=10)

xpos, ypos = np.meshgrid(xedges[:-1], yedges[:-1])
xpos = xpos.flatten()
ypos = ypos.flatten()
zpos = np.zeros_like(xpos)

dx = dy = 0.5 * np.ones_like(zpos)
dz = hist.flatten()

ax.bar3d(xpos, ypos, zpos, dx, dy, dz, color='purple')

ax.set_xlabel("Sepal Length")
ax.set_ylabel("Sepal Width")
ax.set_zlabel("Frequency")

plt.title("3D Histogram")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.linear_model import LinearRegression

# Load dataset
iris = load_iris()

# Independent variable (Sepal Length)
X = iris.data[:, 0].reshape(-1,1)

# Dependent variable (Petal Length)
y = iris.data[:, 2]

# Create regression model
model = LinearRegression()

# Train the model
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# Print coefficients
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)

# Plot regression line
plt.scatter(X, y, color="blue")
plt.plot(X, y_pred, color="red")

plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Linear Regression on Iris Dataset")

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.stats import f_oneway

# Load iris dataset
iris = load_iris()

# Convert to dataframe
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add categorical variable
df['species'] = iris.target

# -------------------------------
# Correlation Matrix
# -------------------------------
corr = df.iloc[:,0:4].corr()

print("Correlation Matrix:")
print(corr)

# -------------------------------
# Correlation Plot (Heatmap)
# -------------------------------
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Plot of Iris Dataset")
plt.show()

# -------------------------------
# ANOVA (Analysis of Variance)
# -------------------------------

setosa = df[df['species']==0]['sepal length (cm)']
versicolor = df[df['species']==1]['sepal length (cm)']
virginica = df[df['species']==2]['sepal length (cm)']

f_stat, p_value = f_oneway(setosa, versicolor, virginica)

print("\nANOVA Result")
print("F-Statistic:", f_stat)
print("P-value:", p_value)
```
